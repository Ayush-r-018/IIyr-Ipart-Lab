```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
```

# Load Data (You can replace this with a real dataset)

```python
data = {
    "Country": ["USA", "India", "Brazil", "Russia", "UK", "Germany", "France", "Italy", "Spain", "Nepal"],
    "Cases": [5000000, 4000000, 3000000, 2500000, 2000000, 1800000, 1600000, 1400000, 1200000, 1000000],
    "Deaths": [250000, 200000, 150000, 120000, 100000, 90000, 80000, 70000, 60000, 50000],
    "Recovered": [4500000, 3700000, 2800000, 2300000, 1800000, 1600000, 1400000, 1300000, 1100000, 950000]
}
```

```python
df = pd.DataFrame(data)
```

# Sort data by cases

```python
df = df.sort_values(by="Cases", ascending=False)
```

# Function to create bar plots

```python
def plot_bar_chart(df, x, y, title, color_palette="coolwarm"):
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x=x, y=y, data=df, palette=color_palette)
    
    for p in ax.patches:
        ax.annotate(f'{p.get_height()/1e6:.1f}M', 
                    (p.get_x() + p.get_width() / 2, p.get_height()), 
                    ha='center', va='bottom', fontsize=12, fontweight='bold')
    
    plt.xlabel(x, fontsize=12)
    plt.ylabel(y, fontsize=12)
    plt.title(title, fontsize=14, fontweight="bold")
    plt.xticks(rotation=45)
    plt.show()
```

# Plot Cases, Deaths, and Recoveries

```python
plot_bar_chart(df, "Country", "Cases", "COVID-19 Cases by Country")
plot_bar_chart(df, "Country", "Deaths", "COVID-19 Deaths by Country", color_palette="Reds")
plot_bar_chart(df, "Country", "Recovered", "COVID-19 Recoveries by Country", color_palette="Greens")
```

# Interactive Plot with Plotly

```python
# Ensure nbformat is installed
fig = px.bar(df, x="Country", y=["Cases", "Deaths", "Recovered"], title="COVID-19 Statistics", barmode='group')
fig.show()
```

# Pie Chart for Death Percentage

```python
df["Death Rate"] = (df["Deaths"] / df["Cases"]) * 100
plt.figure(figsize=(8, 8))
plt.pie(df["Death Rate"], labels=df["Country"], autopct='%1.1f%%', colors=sns.color_palette("Reds", len(df)))
plt.title("COVID-19 Death Rate by Country")
plt.show()
```

# Save the dataframe to CSV

```python
df.to_csv("covid_data.csv", index=False)

print("COVID-19 Data Analysis Completed and Saved as 'covid_data.csv'")
```
