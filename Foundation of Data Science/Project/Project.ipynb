```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Sample COVID data (Replace with real data)

```python
data = {
    "Country": ["USA", "India", "Brazil", "Russia", "UK"],
    "Cases": [5000000, 6000000, 3000000, 2500000, 2000000]
}
```

# Convert to DataFrame

```python
df = pd.DataFrame(data)
```

# Plot Bar Chart

```python
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x="Country", y="Cases", data=df, palette="coolwarm")
plt.xlabel("Country")
plt.ylabel("Cases")
plt.title("COVID-19 Cases by Country")
```

```python
for index, row in df.iterrows():
    barplot.text(index, row.Cases, f'{row.Cases:,}', color='black', ha="center")
    
```

# Labels and Title

```python
plt.xlabel("Countries")
plt.ylabel("Total Cases (in millions)")
plt.title("COVID-19 Cases by Country")
plt.xticks(rotation=45)
plt.grid(True)
```

# Show the plot

```python
plt.show()
```
